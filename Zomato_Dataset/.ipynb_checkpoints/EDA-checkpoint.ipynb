### Zomato Dataset Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv',encoding='iso-8859-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

## Finding if there is a missing value/values in columns

```python
print(df.isnull().sum())
df.isnull().sum().plot(kind = 'bar')
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
country_names = final_df['Country'].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

```python
## plotting the pie chart - Top 3 countries that uses Zomato
plt.pie(x=country_values[:3],labels=country_names[:3],autopct = '%1.2f%%')
plt.show()
```

## Observations: Zomato records maximum transaction records are from India, followed by United States and United Kingdom

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_Count'})
ratings
```

## Observation :: When rating is between 4.5-4.9 ----> Excellent, 4.0-4.4 ------> Very Good, 3.5-3.9 -----> Good, 2.5-3.4 ------> Average, 1.8-2.4 -------> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating_Count', data=ratings)
plt.show()
```

## Observation::
    1. Not Rated count is very high
    2. Maximum number of rating are between 2.5 to 3.4

```python
### Countries that have given 0 rating
set(list(final_df[final_df['Aggregate rating']==0.0]['Country']))
```

```python
## which currency is used by which country
final_df[['Country','Currency']].drop_duplicates()
```

```python
## Which countries have online delivery system ?
set(list(final_df[final_df['Has Online delivery']=='Yes']['Country']))
```

```python
## Citywise distribution of orders
city_values = final_df['City'].value_counts().values
city_labels = final_df['City'].value_counts().index
plt.pie(city_values[:5],labels=city_labels[:5],autopct='%1.2f%%')
plt.show()
```
