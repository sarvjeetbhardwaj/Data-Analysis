```python
df_tesla = pdr.get_data_yahoo('TSLA') #### Getting Tesla share price data
```

```python
#### Plotting the trend of diffrent columns in the dataframe over time
for col in df_tesla.columns:
    df_tesla[col].plot(figsize=(12,4))
    plt.ylabel('Price in $')
    plt.xlabel('Year')
    plt.title(col)
    plt.show()
```

```python
### Visualiing plots in a certain time frame - Setlecting the time frame from Jan 2020 to present date 
for col in df_tesla.columns:
    df_tesla[col].plot(xlim=['2020-01','2022-12'],
                       figsize=(12,4),
                       ls='--')
    plt.ylabel('Price in $')
    plt.xlabel('Year')
    plt.title(col)
    plt.show()
```

### Time Resampling

```python
df_tesla = df_tesla.set_index('Date')
df_tesla
```

```python
df_tesla.resample(rule='A').min() ### Getting the lowest value in each column for every year
```

```python
df_tesla.resample(rule='A').max() ### Getting the maximum values in each column for every year
```

```python
for col in df_tesla.resample(rule='A').max().columns: 
    df_tesla[col].resample(rule='A').max().plot()
    plt.ylabel(f'Maximum {col} values in $ ')
    plt.show()
```

```python
df_tesla.resample(rule='QS').min() ### Getting the lowest value in each column for each quater in an year
```

```python
df_tesla.resample(rule='BA').min() ### Getting the lowest value in each column for each at the end of each Business day in an year
```

```python
for col in df_tesla.resample(rule='A').min().columns:
    df_tesla[col].resample(rule = 'A').min().plot(kind='bar')
    plt.show()
```

```python
df_tesla['Close'].resample(rule ='M').min().plot(kind='bar',figsize=(12,8))
```

## Simple Moving Average

```python
df_tesla['Open:30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,5))
```

```python
for cols in df_tesla.columns:
    df_tesla[f'{cols}:10 days rolling'] = df_tesla[cols].rolling(window=10,min_periods=1).mean()
    df_tesla[[cols,f'{cols}:10 days rolling']].plot(figsize=(12,5))
    plt.title(cols)
    plt.show()
    
```

### Disadvantage of Simple Moving Average is that it gives importance to all the data points.But in timeseries data we give must give more importance to the recent data points 

### Cumulative Moving Average

```python
df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
for cols in df_tesla.columns:
    df_tesla[f'{cols}:Cumulative Moving Average'] = df_tesla[cols].expanding().mean()
    df_tesla[[cols,f'{cols}:Cumulative Moving Average']].plot(figsize=(12,5))
    plt.title(cols)
    plt.show()
    
```

## EMA-Exponential  Moving Average

```python
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust= False).mean()
df_tesla[['Open','EMA_0.1']].plot(figsize=(12,5))
```

```python
df_tesla['EMA_5DAYS'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_5DAYS']].plot(figsize=(12,5))
```
