## Clustering - Univariate, Bivariate, Multivariate

## Univariate Analysis

```python
from sklearn.cluster import KMeans
```

```python
## Determining the appropriate number of clusters using elbow method
inertia_score = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_score.append(kmeans.inertia_)
```

```python
inertia_score
```

```python
plt.plot(range(1,11),inertia_score)
```

```python
## CHOOSING N=3
kmeans = KMeans(n_clusters=3)
kmeans.fit(df[['Annual Income (k$)']])
kmeans.labels_
```

```python
df['Income Cluster'] = kmeans.labels_
df.head()
```

```python
sns.countplot(data=df,x='Income Cluster')
    
```

```python
df.columns
```

```python
df.groupby(['Income Cluster'])['Age', 'Annual Income (k$)', 'Spending Score (1-100)'].mean() ## to check means for each group clusters
```

## Bivariate Clustering

```python
## Determining the appropriate number of clusters using elbow method
inertia_score = []
for j in range(1,11):
    kmeans = KMeans(n_clusters=j)
    kmeans.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
    inertia_score.append(kmeans.inertia_)
inertia_score
    
```

```python
plt.plot(range(1,11),inertia_score)
```

```python
#choosing N=5
clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
df['Spending and Income Cluster'] = clustering2.labels_
```

```python
center = pd.DataFrame(clustering2.cluster_centers_, columns=['x','y'])
```

```python
df.head()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x=center['x'],y=center['y'],s=100,c='black', marker='*')
sns.scatterplot(data=df, x='Annual Income (k$)',y='Spending Score (1-100)',hue='Spending and Income Cluster',palette='tab10')
```

```python
pd.crosstab(df['Spending and Income Cluster'],df['Gender'],normalize='index') ### To get the percentage of male & female in each cluster
```

```python
df.groupby(['Spending and Income Cluster'])['Age', 'Annual Income (k$)', 'Spending Score (1-100)'].mean()
```

## Multivariate Clustering

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar= StandardScaler()
```

```python
dff = pd.get_dummies(df,drop_first=True) ### CPerforming One Hot encoding on the categorical column:Gender
dff.head()
```

```python
dff.columns
```

```python
dff = dff[['Age', 'Annual Income (k$)', 'Spending Score (1-100)','Gender_Male']] ## keeping only the necessary columns
dff.head()
```

```python
dff = pd.DataFrame(scalar.fit_transform(dff)) ## scaling the values using Standard scalar because all the columns are roughly Gaussian
```

```python
dff.head()
```

```python
## Using elbow method 
inertia_score = []
for j in range(1,11):
    kmeans = KMeans(n_clusters=j)
    kmeans.fit(dff)
    inertia_score.append(kmeans.inertia_)
plt.plot(range(1,11),inertia_score)

```

```python
## Choosing N=4
clustering3 = KMeans(n_clusters=4)
clustering3.fit(dff)
df['Clusterig Multivariate'] = clustering3.labels_
df.head()
```

```python
#Nothing meaningful
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='Annual Income (k$)',y='Age',hue='Clusterig Multivariate',palette='tab10')
```

```python
## Below multivariate clustering same as bivariate
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='Annual Income (k$)',y='Spending Score (1-100)',hue='Clusterig Multivariate',palette='tab10')
```
